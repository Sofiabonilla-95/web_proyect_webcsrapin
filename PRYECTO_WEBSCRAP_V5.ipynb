```python
import time 
import re
import sys
import requests 
from pathlib import Path

import numpy as np 
import pandas as pd

from bs4 import BeautifulSoup

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import Select

def url_page():
    path = r'C:\Program Files\ChromeDriver\chromedriver.exe'
    url = 'https://www.ultimatetennisstatistics.com/'
    driver = webdriver.Chrome(path)
    driver.get(url)
    page_source = driver.page_source
    inpute = driver.find_element_by_xpath('//*[@id="menuNavbar"]/ul/li[7]/a')
    inpute.click()
    time.sleep(3)
    leader = driver.find_element_by_xpath('//*[@id="menuNavbar"]/ul/li[7]/ul/li[1]/a/span')
    leader.click()
    driver.refresh()
    inpute = driver.find_element_by_xpath('//*[@id="menuNavbar"]/ul/li[7]/a')
    inpute.click()
    time.sleep(3)
    leader = driver.find_element_by_xpath('//*[@id="menuNavbar"]/ul/li[7]/ul/li[1]/a/span')
    leader.click()
    
def seleccionar_categoria(categoria):
    lista_des= Select(driver.find_element_by_id('category'))
    return lista_des.select_by_visible_text(categoria)
       
def imprimir_tabla():
    lista=[]
    for i in range(1, 16):
        page_2= driver.find_element_by_xpath(f'//a[@data-page="{i}"]')
        page_2.click()
        time.sleep(5)
        texto_2 = driver.find_elements_by_tag_name('tr')
        texto=[link.text.strip() for link in texto_2]
        lista.append(texto)
    return lista

def arreglar_info(lista):
    matriz=[y for x in lista for y in x]
    del matriz[0:300:21]
    matriz_1=[]
    for i in range(len(matriz)):
        a=matriz[i].split()
        matriz_1.append(a)
    matriz_2=[x for x in matriz_1]
    for y in range(len(matriz_2)):
        if len(matriz_2[y])>5:
            del matriz_2[y][3]
        else:
            matriz_2[y]
    matriz_3=matriz_2
    for y in range(len(matriz_3)):
        if len(matriz_3[y])>5:
            del matriz_3[y][3]
    df = pd.DataFrame(matriz_3)
    df.columns = ['Rank', 'Country', 'Name','Last name','Matches Played']
    df.insert(loc = 4, 
          column = 'Full Name', 
          value = df["Name"] + " " + df["Last name"]) 
    return df

def insertar_columna(lista,tabla):
    matriz=[y for x in lista for y in x]
    del matriz[0:300:21]
    matriz_1=[]
    for i in range(len(matriz)):
        a=matriz[i].split()
        matriz_1.append(a)
    matriz_2=[x for x in matriz_1]
    for y in range(len(matriz_2)):
        if len(matriz_2[y])>5:
            del matriz_2[y][3]
        else:
            matriz_2[y]
    matriz_3=matriz_2
    for y in range(len(matriz_3)):
        if len(matriz_3[y])>5:
            del matriz_3[y][3]
    df = pd.DataFrame(matriz_3)
    df.columns = ['Rank', 'Country', 'Name','Last name',tabla]
    df.insert(loc = 4, 
          column = 'Full Name', 
          value = df["Name"] + " " + df["Last name"]) 
    dfb=df.iloc[:,-2:]
    return dfb

def data_frame(a,b):
    c=pd.merge(a, b, how="left", on=["Full Name"])
    return(c)

#url_page()
#url = 'https://www.ultimatetennisstatistics.com/#google_vignette'

#ITRERAR

path = r'C:\Program Files\ChromeDriver\chromedriver.exe'
url = 'https://www.ultimatetennisstatistics.com/'
driver = webdriver.Chrome(path)
driver.get(url)
page_source = driver.page_source
inpute = driver.find_element_by_xpath('//*[@id="menuNavbar"]/ul/li[7]/a')
inpute.click()
time.sleep(3)
leader = driver.find_element_by_xpath('//*[@id="menuNavbar"]/ul/li[7]/ul/li[1]/a/span')
leader.click()
time.sleep(5)

seleccionar_categoria('Matches Played')
time.sleep(3)
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
data=imprimir_tabla()
arreglar_info(data)
time.sleep(3)

seleccionar_categoria('Matches Won')
time.sleep(3)
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
time.sleep(5)
data_1=imprimir_tabla()
insertar_columna(data_1,'Matches Won')
time.sleep(3)

seleccionar_categoria('Match Time')
time.sleep(3)
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
time.sleep(5)
data_2=imprimir_tabla()
insertar_columna(data_2,'Match Time')
time.sleep(3)

seleccionar_categoria('Aces')
time.sleep(3)
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
time.sleep(5)
data_3=imprimir_tabla()
insertar_columna(data_3,'Aces')
time.sleep(3)

seleccionar_categoria('Double Faults')
time.sleep(3)
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
time.sleep(5)
data_4=imprimir_tabla()
insertar_columna(data_4,'Double Faults')
time.sleep(3)

seleccionar_categoria('1st Serve Won %')
time.sleep(3)
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
time.sleep(5)
data_5=imprimir_tabla()
insertar_columna(data_5,'1st Serve Won %')
time.sleep(3)

seleccionar_categoria('2nd Serve Won %')
time.sleep(3)
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
time.sleep(5)
data_6=imprimir_tabla()
insertar_columna(data_6,'2nd Serve Won %')
time.sleep(3)

seleccionar_categoria('Points per Match')
time.sleep(3)
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
time.sleep(5)
data_7=imprimir_tabla()
insertar_columna(data_7,'Points per Match')
time.sleep(3)

seleccionar_categoria('Break Points Saved %')
time.sleep(3)
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
time.sleep(5)
data_8=imprimir_tabla()
insertar_columna(data_8,'Break Points Saved %')
time.sleep(3)

seleccionar_categoria('Tie-Breaks Won %')
time.sleep(3)
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
time.sleep(5)
data_9=imprimir_tabla()
insertar_columna(data_9,'Tie-Breaks Won %')

#armar matriz

a=data_frame(arreglar_info(data),insertar_columna(data_1,'Matches Won'))
b=data_frame(a,insertar_columna(data_2,'Match Time'))
c=data_frame(b,insertar_columna(data_3,'Aces'))
d=data_frame(c,insertar_columna(data_4,'Double Faults'))
e=data_frame(d,insertar_columna(data_5,'1st Serve Won %'))
f=data_frame(e,insertar_columna(data_6,'2nd Serve Won %'))
g=data_frame(f,insertar_columna(data_7,'Points per Match'))
h=data_frame(g,insertar_columna(data_8,'Break Points Saved %'))
i=data_frame(h,insertar_columna(data_9,'Tie-Breaks Won %'))

driver.quit()
```
